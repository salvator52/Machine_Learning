# Sınıflandırma

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("maliciousornot.xlsx")
df #548 satır 31 sütun var
```

```python
df.describe()
```

```python
sns.countplot(x="Type",data=df)
```

```python
df.corr()["Type"].sort_values().plot(kind="bar")
```

```python
df_copy = df.copy()
y = df_copy["Type"].values
df_copy.drop("Type",axis=1,inplace=True)
X = df_copy.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_X_train = scaler.fit_transform(X_train)
scale_X_test = scaler.fit_transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))

#sigmoid sınıflandırma problemlerinde çıkış layer'ına koyulabilir
model.add(Dense(units=1,activation="sigmoid"))

#binary classification olduğu için loss'a yeni bir değer koyduk
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(
    x=scale_X_train, 
    y=y_train,
    validation_data=(scale_X_test, y_test),
    verbose=1,
    epochs=700
)
```

```python
lossData = pd.DataFrame(model.history.history)
lossData.plot() #epoch 700 olarak kullanıldığında model overfitting yaptığını görebiliriz
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))

#sigmoid sınıflandırma problemlerinde çıkış layer'ına koyulabilir
model.add(Dense(units=1,activation="sigmoid"))

#binary classification olduğu için loss'a yeni bir değer koyduk
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
#early stopping
#İstediğimiz kadar epoch değeri versekte belli bir düzeyi(bizim belirlediğimiz) kendi otomatik olarak durdurur ve overfitting'in önüne geçer
#monitor İzlenecek metriği belirtir
#verbose Erken durdurma olayları hakkında bilgi vermek için kullanılır
#patience zlenen metriğin belirlenen monitor değeri iyileşmeden kaç epoch (eğitim döngüsü) geçtikten sonra erken durdurmanın uygulanacağını belirtir.
earlyStopping = EarlyStopping(monitor = "val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(
    x=scale_X_train, 
    y=y_train, 
    epochs=700, 
    validation_data=(scale_X_test, y_test), 
    verbose=1, 
    callbacks=[earlyStopping]
)
```

```python
lossData = pd.DataFrame(model.history.history)
lossData.plot()
```

```python
#Dropout verilen yüzdeyi o kadar layer'ı rastgele atacak,layerlarla alakalı overfitting yaşıyorsak
#bunu engellicek ve daha iyi bir sonuç sağlayabilir 0-1 arası değer alır
```

```python
model = Sequential()
#her katmandan sonraya dropout koyalım
model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0,6)) #0.5 üstü uyarı verir
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0,6))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0,6))
#sigmoid sınıflandırma problemlerinde çıkış layer'ına koyulabilir
model.add(Dense(units=1,activation="sigmoid"))

#binary classification olduğu için loss'a yeni bir değer koyduk
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(
    x=scale_X_train, 
    y=y_train, 
    epochs=700, 
    validation_data=(scale_X_test, y_test), 
    verbose=1, 
    callbacks=[earlyStopping]
)
```

```python
lossData = pd.DataFrame(model.history.history)
lossData.plot()
#label overfitting'den kurtardık ve grafiğimiz daha birbirine benzer bir hal aldı
```
